```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
df=df[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed',
               'Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
df.head()
```

```python
df['Country'].value_counts().head()
```

```python
print("Country with the most attacks:", df['Country'].value_counts().idxmax())
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df['Country'].value_counts()[:15].index,y=df['Country'].value_counts().values[:15], palette = 'rocket')
plt.title('Countries that got attacked the most')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('Attacks')
plt.show()
```

```python
df['Country'].value_counts().head() 
df['Country'].value_counts().head().plot(kind = 'pie', figsize = (20,8), fontsize=10, autopct= '%3.1f%%', 
                                                shadow=True, startangle=135, legend=True, )
plt.ylabel("Countries", size = 10)
plt.xlabel("Attacks Occured", size = 10)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
df['Region'].value_counts().head() 
```

```python
df['Region'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=df['Region'].value_counts()[:14].index,y=df['Region'].value_counts().values[:14], palette = 'deep')
plt.title('Regions that  got attacked the most')
plt.xticks(rotation=90)
plt.xlabel('Regions')
plt.ylabel('Attacks')
plt.show()
```

```python
df['Region'].value_counts().head() 
df['Region'].value_counts().head().plot(kind = 'pie', figsize = (15,8), fontsize=10, colors = ['#ff6666', '#ffcc99', '#99ff99', '#66b3ff'], 
                                                shadow=True, startangle=135, legend=True, )
plt.ylabel("Regions", size = 10)
plt.xlabel("Attacks Occured", size = 10)
plt.xticks(size = 10)
plt.yticks(size = 10)
```

```python
df['Region'].value_counts().tail().plot(kind = 'pie', figsize = (20,8), fontsize=10, colors = ['#ff6666', '#ffcc99', '#99ff99', '#66b3ff'], 
                                                shadow=True, startangle=135, legend=True, )
plt.ylabel("Regions", size = 10)
plt.xlabel("Attacks Occured", size = 10)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
df['city'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=df['city'].value_counts()[:15].index,y=df['Country'].value_counts().values[:15])
plt.title('Cities that are  attacked the most')
plt.xticks(rotation=90)
plt.xlabel('Cities')
plt.ylabel('Attacks')
plt.show()
```

```python
df.groupby('Year')['Killed'].sum().plot(kind = 'line', linewidth = 5, figsize =(15,5),color = 'red')
plt.title("Trend of deaths due to terror attacks",fontsize= 10, fontweight = 'bold')
plt.xlabel("Years", size = 16)
plt.ylabel("Total deaths occured ", size = 10)
plt.xticks(size = 10)
plt.yticks(size = 10)
```

```python
df['AttackType'].value_counts().head() 
df['AttackType'].value_counts().head().plot(kind = 'pie', figsize = (18,8), fontsize=10, autopct= '%3.1f%%', 
                                                shadow=True, startangle=135, legend=True, )
plt.ylabel("Countries", size = 10)
plt.xlabel("Attacks Occured", size = 10)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
df['AttackType'].value_counts().tail() 
df['AttackType'].value_counts().tail().plot(kind = 'pie', figsize = (18,8), fontsize=10, autopct= '%3.1f%%', 
                                                shadow=True, startangle=135, legend=True, )
plt.ylabel("Countries", size = 10)
plt.xlabel("Attacks Occured", size = 10)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=df['Group'].value_counts()[:20].index,y=df['Group'].value_counts()[:20].values,palette='colorblind')
plt.xticks(rotation=90)
plt.title('Total Group Attacks')
plt.xlabel('Group Name')
plt.ylabel('Total no. of attacks')
plt.show()
```

```python
df['Group'].value_counts().tail() 
df['Group'].value_counts().tail().plot(kind = 'pie', figsize = (15,8), fontsize=10, colors = ['#ff6666', '#ffcc99', '#99ff99', '#66b3ff'], 
                                                shadow=True, startangle=135, legend=True, )
plt.ylabel("Group", size = 10)
plt.xlabel("Attacks Occured", size = 10)
plt.xticks(size = 10)
plt.yticks(size = 10)
```
